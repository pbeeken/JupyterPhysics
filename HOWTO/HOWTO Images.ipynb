# How to import images and diagrams
Embedding homemade images into Jupyter is straight forward if the files reside on your own computer but doing this with gDrive stored images can be tricky. If the image is not hosted on a server and made public it is tough to present images you have created in your own Google Drive folder without understanding how to make them public and how to access them as if they were on a public server.  This sheet is a set of examples to get something like this to happen.

As with other HOWTO guides this document is meant to be interactive. Make a copy of it and try the things presented here and let it become your own reference.

## Preamble

Images, like all "files" on Google Drive aren't files in the traditional sense. There isn't a folder hierarchy and the files aren't physical regions of storage where their information is stored. Instead they are 'blobs' (an actual technical term) of data indexed in a huge database that spans the world.  The real name of your file is a long meaningless string of numbers called an ID.  The name you see is just for your convenience.  

The ability to make these blobs visible and then get them to resolve to a file type the browser will recognize requires 2 steps.  With traditional websites all you have to do is to refer to the folder hierarchy by name and eventually to the file. The file is passed back to the browser along with a message that it is a particular type (an image of kind jpg, for example) so the browser knows what to do with it.  

gDrive is a database with blobs. When a request comes from a source like a Notebook (or webpage, for that matter) gDrive just hands it back as the blob it thinks it is. Changing what we call the image can change how gDrive hands it back which is why you have to **A** make the image public so the data can be passed back and **B** change the URL in subtle but important ways so the receiver knows it is an image and of what type.

## Directly embedding
This is a trick where the data of the image is copied directly into a cell (text or code) so if the notebook gets moved the data for the image follows the document.  This can be cumbersome as large image files can take up a great deal of space.  For <abbr title="png, jpg and gif">small images</abbr> and <abbr title="svg images">vector graphical images</abbr> this can be the easiest way to embed an image.

#### **base64** encoding

In CoLab, if you copy an image and then 'paste' it into a text cell it is pasted in a format that is interpreted as *base64* encoded data as a standard markdown image. These kinds of images are called pixel images which means they are like paintings where colors are laid down in a drawing. The base64 encoding is an old school way of portraying a binary file in a way that can be copied in a plain text editor.  It is commonly used in email messages. For small png files this is a universally transportable method to include a small diagram or icon in a text field.

### Preparing the graphics file for presentation

#### Using **Google Drawings** App


The easiest way to make an image or drawing available to a notebook is to create a 'drawing' document in your gDrive.  Either create the drawing from scratch with the tools available there or paste the desired image file.  When you are done shrink the drawing area to the size of the image (drag the chevrons from the lower right corner to close the drawing space down).

Make the image publicly visible for viewing.  There are three steps involved in making an image available to a jupyter notebook. N.B. this works reliably in CoLab and should work in any jupyter implementation but hasn't been tested as extensively.

  1. Click on the share button and allow anyone with the link to view$^\ddagger$.  (*You don't need to make it fully publicly available*)
  2. Copy the URL using the `Get Link` dialog (*easiest way*)
  3. Paste the URL into a <abbr title="![alternate title](URL)">markdown tag</abbr> or an <abbr title='<img src="URL".../>'>html tag</abbr> then edit the ending.  
  > The URL should look something like this:
  > `...docs.google.com/drawings/d/*fileID*/edit?usp=sharing`
  >
  > replace the ending `/edit?usp=sharing` with either `/export/svg` or `/export/png` and the image should appear in your rendered cell.


---
$^\ddagger$<small>With enterprise accounts (as those with schools and universities) you can also make things visible to everyone within an enterprise but not widely available. As evidenced with tests done below this is insufficient.  This makes sense. Your personal account may have a gDrive privilege to look at your own files but a virtual machine runnning in the cloud doesn't.  Wihout that connection (like a Jupyter kernel running on your own machine or server) the file has to be accessible publicly.</small>

````
#@title Convert Drawing file link to image url
#@markdown Get the link from the 'share' button or menu item... 
raw_url = "https://docs.google.com/drawings/d/somewild_andcrazycharacters/edit?usp=sharing"  #@param {type: "string"}
output = "svg" #@param ["png", "svg"]
#@markdown copy one of the following lines and paste into your text cell
conv_url = raw_url.replace("/edit?usp=sharing", "/export/" + output)
print(f'![alt text]("{conv_url}")')
print("or")
print(f'<img alt="alt text" src="{conv_url}" align="right" width="200px"/>')
````

### Presenting using **HTML**
At this point you have followed one of the two steps above to fashion the URL to the image you want to present.

HTML works for your own files as well as it does for public images.  With a little bit of extra overhead you can control the image size and positioning.
The html tag looks like this

`<img src="`*URL*`" align="`*LOC*`" height="`*HGT*`" width="`*WID*`" border="`*SZE*`"/>`
  > *URL* [*required*] is the URL you have pasted from your drawing or image and modified according to instructions.
  >
  > *LOC* [*optional*] can be RIGHT or LEFT and is the location you want the image to drift text will flow around the image on the opposite side.
  >
  > *HGT*, *WID* [*optional*] use neither, either, or both. The value is the height or width in pixels. If you use neither then the image is original size, setting either height or width will scale that side keeping the objects aspect ratio the same. Using both may distort the image to fit scale in both dimensions.
  >
  > *SZE* [*optional*] is a black border of SZE width drawn around the image, the default is no border.
