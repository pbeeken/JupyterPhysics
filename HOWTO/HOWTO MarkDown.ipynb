### Indentation

Lists can be made in much the way that you might if you were typing in a plain text editor. A List is started by simply indenting a line with 2 spaces and typing...
  * a `*` or `-` for an unordered list (it doesn't matter which)
    - you can mix and match (notice you need to increase the indent with 2 spaces.) 
    - but it doesn't matter which symbol you use.
  * a number (1,2,3...) followed by a period (e.g. 1.) for an ordered list.
  * a greater-than symbol for a straight unmarked indent `>`.
    - Yes, you can nest them to increase the indent.

An ordered list is a sequence of numbers (*the increment is automatic so your numbers don't need to be sequential*) followed by a period:`.`
  1. first number sets the start of the count.
  5. second number doesn't matter as long as it is a number
  0. third 
  1000. Notice that the number after the first doesn't matter for subsequent values.

Indents are often used for quotes or captions that need special attention.
    
  > *What doth gravity out of his bed at midnight?*
  >> *The labor we delight in physics pain.*

### Tables

Tables are done by surrounding the elements with vertical bars `|`
  - first line is the title and sets the number of columns
  - second line defines the justification (left, right or center)
    - `---` or `:--` left justifies the elements in their column (default)
    - `--:` right justifies
    - `:-:` centers the elements
  - subsequent lines are the row elements with each element separated by `|`
    - if you want $n$ coulumns then there should be $n+1$ vertical bars on each line.
    - e.g for 3 columns there should be 4 vertical bars

|item| x [*m*] | F [*N*] |
|:-:|:--|--:|
|a|0.3| 12.|
|b|0.4|1|
|c|0.5|2.3E2|
|d|1.2|3,200|

## Equations and formulas 
  > *a.k.a.* $\LaTeX$ formatting

In physics, engineering, chemistry, and indeed many areas of science and mathematics we need to document our logic with formulas and equations.  **jupyter** markdown makes it easy to embed equations.  Initially you can get help using a huge assortment of [graphical editors](http://atomurl.net/math/) but if you watch closely there is a simple syntax to the formula construction.  The link I provided is one where it generates the required $\LaTeX$ syntax.  Take this example copied straight from this site:
$$\vec{F}=m \frac{d \vec{v}}{dt} + \vec{v}\frac{dm}{dt}$$
Your choice, therefore, is whether to set the formula off on its own line as you see above or embed the expression in the line: $\vec{F} = m \vec{a}$. There are many many tags but learning just a few allows you to write formulas as you type.

Here are some neat examples to  contemplate:
  - An embedded rendition of the pythagorean formula $h^2=x^2+y^2$
      - was generated from "`..agorean formula $h^2=x^2+y^2\$`"
   - more complicated formulas or ones that you want to emphasize can be set off on their own line by embedding the $LaTeX$ betwen a pair of double dollar signs ('\$\$'). *N.B. When you use the double \$\$ get into the habit of labelling the equation with an **_tag_** so you can refer to it in your text.* 
  
  - thus `$$x_{\pm} = \frac{-b \pm \sqrt{b^2-4 a c}}{2 a} \tag{1}$$` becomes...
   $$x_{\pm} = \frac{-b \pm \sqrt{b^2-4 a c}}{2 a} \tag{1}$$
  
  - $\LaTeX$ code itself contains a vast number of possabilities but is very simple at its core. For the most part you type the formula as you might into a calculator.
      - `$y = m x + b$` : $y = m x + b$
      - `$x(t) = x_0 + v_0 t + \frac{1}{2} a t^2\$` : $x(t) = x_0 + v_0 t + \frac{1}{2} a t^2$
        - subscripts are rendered after an underbar '_' and superscripts are rendered after a caret '^'
        - notice the special construct for fractions: \frac{numerator}{denominator}.
        - In general, curly braces are used as grouping multiple symbols.
  
  - There are a rich set of special characters like greek letters and special symbols:
     - \alpha, \beta, \gamma, or \Delta vs \delta 
     - $\alpha$, $\beta$, $\gamma$, or $\Delta$ vs $\delta$
     - You can even format vectors $\vec{x}$ and $\vec{F}$ so you can refer to these items properly.
     - Here's a neat trick for fractions: ` $^3 / _{18}$ ` becomes $^3 / _{18}$
     - How about drawing a box around an answer: $\boxed{3.43 J}$

there are many constructs to create formulas, you can [search for](https://www.google.com/search?q=latex+formula+cheat+sheet&oq=latex+formula&aqs=chrome.2.69i57j0l5.8175j0j4&sourceid=chrome&ie=UTF-8) many different takes on this. #### Aligning Formulas

You want to document a derivation which usually requires multiple steps.  We usually present these as a series of lines with the equal sign lining up:
  - Start with a \\begin{align} ... end{align} pair
  - Each step will separate the <abbr text="left hand side">lhs</abbr> from the the <abbr text="right hand side">rhs</abbr> with a `& =` ending with `\\` \
    (lhs &= rhs \\\\)

$$
\begin{align}
a_c & = \omega^2 r \tag{3}\\
    & = \left( \frac{v}{r} \right)^2 r\\
    & = v^2\left( \frac{r}{r^2} \right)\\
    & = \frac{v^2}{r} \tag{4}
\end{align}
$$



#### Crossing out

Suppose you want to cross out parts of a formula as part of a derivation:
$$q\; v \; B = m \frac{v^2}{r}$$
simplifying$\dots$
$$\require{cancel}   q\; \cancel{v} \; B = m \frac{v^\bcancel{2}}{r}$$
solving for r$\dots$
$$\require{enclose}   \frac{r}{\enclose{horizontalstrike}{q\;B}}\quad \enclose{horizontalstrike}{q\; B} = \frac{\cancel{r}}{q\;B}\quad m \frac{v}{\cancel{r}}$$
simplifying$\dots$
$$\bbox[yellow]{r = \frac{m v}{q\;B}}$$

#### Formating Numbers and Units

We display alot of numbers with units in physics.  In order to insure readability and consistency we need to use the following formats.

Rule 1: **SIGNIFICANT DIGITS!** Limit the number of significant digits. If you measurement and/or calculation involves ony 3-4 sig. fig. then your answer cannot exceed this.  All the examples require that you know how to an do limit your answers.

$$\require{cancel}\cancel{2.3453424832764032}\tag{✘}$$
$$2.345\tag{✔}$$

Rule 2: **APPROPRIATE REPRESENTATION**  Really big and small numbers can be presented one of two ways: using scientific notation or using prefixes.  

Rule 3: **UNITS!** All measurements in physics have units and only in rare circumstances are values dimensionless.  Format the units as you would for any algebraic expression...

$$3.64\cdot 10^{10}\,J\tag{sci. not.}$$
$$36.4\,GJ\tag{prefix}$$

Compound units can be composed with the same tricks for algebra (here we include a fancy trick to squeeze the characters together a bit using the negative space seperator `\!`
$$3.64\cdot 10^{10}\, N \! \cdot \! m\tag{sci. not.}$$

if the unit involves a fraction there are two ways to do it...
$$3.64\cdot 10^{-2}\,\frac{m}{s^2}\tag{sci. not.}$$

or inline where the negaitve space `\!` is used very effectively:
$$3.64\cdot 10^{-2}\,m\! / \! s^2\tag{sci. not.}$$

A stack of various examples:

$$180^\circ=\pi \, ^{\scriptsize{rad}} $$

$$N_A = 6.022\times10^{23} \ \mathrm{mol}^{-1}$$
