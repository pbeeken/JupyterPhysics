Conservation of the forces that act within a "*system*".  The notion of conservation laws of any kind involves the idea of seperation. We can isolate a group of objects: a set of atoms, a set of molecules, a pile of rocks, a solar system and call that a system and everything else becomes the '*universe*'.  With this seperation we isolate the things that are '*group*' and consider the forces that act internally creating a whole where all the interactions preserve the momentum of within so that whatever energy and momentum contained within stays within.  Forces that act on the '*group*' from without (from the '*universe*') are the only way the system can be changed.  This is the way we can see how the microscopic becomes the macroscopic.


## Linear Momentum
This is about following what happens to colliding objects interacting in a straight line. (I am sure you wont be surprised that things generally interact in 3 dimensions) Still we start with spherical cows to give us insight into more complex interactions.

We start with two assertions (ones we have very good reason to accept): Momentum ($\vec{p}$) and Energy (*E*) are conserved. Momentum has only one form but Energy can take on 3 forms ($K$) kinetic, ($U$) potential, and ($Q$) heat.  As mentioned when we visited the Energy concept $K$ (kinetic) and $U$ (potential) are often coupled together and called "Mechanical" energy. $Q$ (heat) is more complex because of what we to do to convert it back to mechanical energy.

So lets consider two options.  Interactions where the collisions preserve the kinetic energy the ones that do not.

These interactions all have two states: a before and an after.  We can infer the "while" from the difference between these two limits.  Consider the simplest case two objects:  object 1: $m_1$, $v_{1b}$, $v_{1a}$ and object 2: $m_2$, $v_{2b}$, $v_{2a}$.
These are quite a mouthful of variables but we can make one simplification.  We have already convinced ourselves that a "Galilean Transformation" does not alter Newton's Laws. We simply shift our frame of reference by changing adding or subtracting a uniform velocity from all of the objects. (If we took this whole experiment and put in on a train moving at uniform speed on an ultra smooth track nothing would change.) Let's subtract $v_{1b}$ from all the above velocities so the initial collision has the second object standing still with $v_{2b}=0$.  To keep things simple we will simply eliminate $v_{2b}$

In general:

$$
\begin{align}
p_b  &= p_a\\
m_1 \; v_{1b}   &=  m_1 \; v_{1a} + m_2 \; v_{2a}\tag{momentum}\\
E_b &= E_a \tag{total energy}\\
\end{align}
$$
This is all we can do with this combination except to consider different ways the energy can be transformed: Perhaps $E_b$ is all potential energy that is transformed into kinetic energy. Maybe we run the equations backwards and as the two moving carts come together I trap the kinetic energy into some kind of storage system or disipate it as heat?

But we can then consider a special case where kinetic energy is conserved:
$$
\begin{align}
p_b  &= p_a\\
m_1 \; v_{1b}  &=  m_1 \; v_{1a} + m_2 \; v_{2a}\tag{momentum}\\
K_b &= K_a \tag{kinetic energy}\\
\frac{1}{2}m_1 \; v_{1b}^2   &=  \frac{1}{2}m_1 \; v_{1a}^2 + \frac{1}{2}m_2 \; v_{2a}^2 \tag{kinetic energy}\\
\end{align}
$$

and ask if we can solve these two independent equations and two unknowns. All the masses and the inital velocity $v_{1b}$ are known.

The first step might be to eliminate the $v_2a$ from the two unknown variables.  

$$
\begin{array}{lll}
m_1 \; v_{1b}  =  m_1 \; v_{1a} + m_2 \; v_{2a} \implies &v_{2a}  =  \frac{m_1}{m_2} \; v_{1a} - \frac{m_1}{m_2} \; v_{1b}  \\
\frac{1}{2}m_1 \; v_{1b}^2  =  \frac{1}{2}m_1 \; v_{1a}^2 + \frac{1}{2}m_2 \; v_{2a}^2 \implies &\frac{1}{2}m_1 \; v_{1b}^2 = \frac{1}{2}m_1 \; v_{1a}^2 + \frac{1}{2}m_2 \; \left( \frac{m_1}{m_2} \; v_{1a} - \frac{m_1}{m_2} \; v_{1b} \right)^2 \\
&m_1 \; v_{1b}^2 = m_1 \; v_{1a}^2 + m_2 \; \frac{m_1^2}{m_2^2} \left( v_{1a} - v_{1b} \right)^2  \\
& v_{1b}^2 - v_{1a}^2 = \frac{m_1}{m_2} \left(-\left[ v_{1b} - v_{1a} \right] \right)^2  \\
& \left(v_{1b} - v_{1a}\right)\;\left(v_{1b} + v_{1a}\right) = \frac{m_1}{m_2} \left(v_{1b} - v_{1a}  \right)^2  \\
& \left(v_{1b} + v_{1a}\right) = \frac{m_1}{m_2} \left(v_{1b} - v_{1a}  \right)  \\
& m_2\;\left(v_{1b} + v_{1a}\right) = m_1 \left(v_{1b} - v_{1a}  \right) \\
& \left(m_2 - m_1\right)\;v_{1b} = -\left(m_1+m_2\right) v_{1a} \\
& \left(m_1 - m_2\right)\;v_{1b} = \left(m_1+m_2\right) v_{1a} \\
& v_{1a} = \frac{m_1 - m_2}{m_1+m_2}\;v_{1b}  \\
\end{array}
$$

To solve for the second cart's velocity simply substitute into the momentum equation and solve for $v_{2a}$.


```python
#@title Simulation
import numpy as np
import matplotlib.pyplot as plt

default_simulation_controls = {
    "timeLimit": 200,    # time to stop simulation
    "timeStep":  0.1,    # time step
    "start_x_1":   5,    # x position of first object
                         # y position replaced by impactParameter/2
    "start_x_2":  -5,    # x position of second object
                         # y position replaced by impactParameter/2
}

# This allows us to play with different scattering potentials
def interactionForce(m1, m2, r):
    """
    Determines the interaction force between two objects of mass m1 and m2
    m1 and m2 are simply the means to scale the strength of the force between the objects
    r is the distance between the objects, always >= 0
    returns the force between the objects
    """
    return - 0.4 * m1 * m2 / r**6
    # return - m1 * m2 * np.exp(-r/100)

# Function to simulate the collision
def collision_simulation(m1, m2, v_initial, impact_parameter, simulation_controls=default_simulation_controls):
    """
    Simulates the collision of two objects with a given initial velocity and impact parameter.
    m1 and m2 are the mediators for the interaction strength
    v_initial is the initial velocity of the objects as they approach one another.
    impact_parameter is the distance of seperation between as they approach
    """
    # Initial conditions
    x1 = simulation_controls["start_x_1"]
    y1 = + impact_parameter / 2
    x2 = simulation_controls["start_x_2"]
    y2 = - impact_parameter / 2

    vx1 = -v_initial
    vy1 = 0
    vx2 = +v_initial
    vy2 = 0

    # Time step and simulation duration
    dt = simulation_controls['timeStep']
    t_max = simulation_controls['timeLimit']

    # Lists to store trajectory data
    x1_history = [x1]
    y1_history = [y1]
    x2_history = [x2]
    y2_history = [y2]

    for t in np.arange(0, t_max, dt):
      # Calculate the distance between the objects
      distance = np.hypot(x2 - x1, y2 - y1)
    #   distance = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

      if distance > 0.1: # Avoid division by zero and simulate free movement when close together
        # Calculate the gravitational force
        force = interactionForce(m1, m2, distance)

        # Calculate the angle between the objects
        angle = np.arctan2(y2 - y1, x2 - x1)

        # Calculate the force components
        fx = force * np.cos(angle)
        fy = force * np.sin(angle)

        # Update velocities
        vx1 += fx / m1 * dt
        vy1 += fy / m1 * dt
        vx2 -= fx / m2 * dt
        vy2 -= fy / m2 * dt

        # Update positions
        x1 += vx1 * dt
        y1 += vy1 * dt
        x2 += vx2 * dt
        y2 += vy2 * dt
      else: # If distance between objects becomes very small, set equal velocity to simulate collision effect
        vx1 = vx2 = (vx1*m1 + vx2*m2) / (m1 + m2)
        vy1 = vy2 = (vy1*m1 + vy2*m2) / (m1 + m2)

      # Add the current positions to the history
      x1_history.append(x1)
      y1_history.append(y1)
      x2_history.append(x2)
      y2_history.append(y2)

    m1_history = np.array([x1_history,y1_history]).T
    m2_history = np.array([x2_history,y2_history]).T
    # Return position history
    return m1_history, m2_history

# Uniform way to display trajectories
def plotTrjectories(obj1_history, obj2_history, ax=None):
    """
    Uniform way to plot the trajectories of the objects.
    TODO: provide a way to work from a axes object
    """
    if ax is None:
        fig, ax = plt.subplots()

    [obj1_x, obj1_y] = obj1_history.T
    [obj2_x, obj2_y] = obj2_history.T

    # Plot the trajectories
    ax.plot(obj1_x, obj1_y, label='Puck 1')
    ax.plot(obj2_x, obj2_y, label='Puck 2')
    ax.set_xlabel('X Position')
    ax.set_ylabel('Y Position')
    ax.set_title('Collision Simulation')
    ax.legend()
    ax.grid(True)
    ax.set_box_aspect(1)
    ax.set_ylim(-10, 10)
    ax.set_xlim(-10, 10)
    ax.text(-9.5,-9.5, f"ip: {obj1_y[0]-obj2_y[0]}")
    # plt.show()

def plotSpeeds(obj1_history, obj2_history, ax=None):
    """
    Uniform way to plot the trajectories of the objects.
    TODO: provide a way to work from a axes object
    """
    if ax is None:
        fig, ax = plt.subplots()

    [obj1_x, obj1_y] = obj1_history.T
    [obj2_x, obj2_y] = obj2_history.T

    spd1 = np.hypot(np.diff(obj1_x), np.diff(obj1_y)) / 0.1
    spd2 = np.hypot(np.diff(obj1_x), np.diff(obj1_y)) / 0.1
    r = np.hypot(obj1_x, obj1_y)[:-1]

    ax.plot(r, spd1, c="blue", alpha=0.3, linestyle="dotted", linewidth=4, label="Puck 1")
    ax.plot(r, spd2, c="red",  alpha=0.3, linestyle="dashed", linewidth=4, label="Puck 2")
    ax.set_xlabel('distance')
    ax.set_ylabel('velocity')
    ax.legend()
    ax.grid(True)

# Example usage
m_1 = 1
m_2 = 2
v_0 = 0.05
impact_parameter = 1.2

fig, ((a1, a2),(v1, v2)) = plt.subplots(2, 2)
fig.set_size_inches(10, 10)
plt.subplots_adjust(wspace=0.2, hspace=0.3) # left=0.1, bottom=0.1, right=0.9, top=0.9,

obj1, obj2 = collision_simulation(m_1, m_2, v_0, impact_parameter)
plotTrjectories(obj1, obj2, a1)
plotSpeeds(obj1, obj2, v1)

obj1, obj2 = collision_simulation(m_1, m_2, v_0, np.round(impact_parameter*1.5,2))
plotTrjectories(obj1, obj2, a2)
plotSpeeds(obj1, obj2, v2)

plt.show()
```

```python
def phasePlot(obj1_history, obj2_history):
    [obj1_x, obj1_y] = obj1_history.T
    [obj2_x, obj2_y] = obj2_history.T

    vel1_x = np.diff(obj1_x) / 0.1
    vel1_y = np.diff(obj1_y) / 0.1
    vel2_x = np.diff(obj2_x) / 0.1
    vel2_y = np.diff(obj2_y) / 0.1

    pos1_x = obj1_x[:-1]
    pos1_y = obj1_y[:-1]
    pos2_x = obj2_x[:-1]
    pos2_y = obj2_y[:-1]

    plt.plot(pos1_x, vel1_x, c="red",  label='Puck 1 x')
    plt.plot(pos1_y, vel1_y, c="magenta", label='Puck 1 y')

    plt.plot(pos2_x, vel2_x, c="blue",  label='Puck 2 x')
    plt.plot(pos2_y, vel2_y, c="cyan", label='Puck 2 y')

    plt.xlabel('position')
    plt.ylabel('velocity')

    plt.title('Phase Space')

    plt.legend()
    plt.grid(True)
    plt.show()

phasePlot(obj1, obj2)
```
