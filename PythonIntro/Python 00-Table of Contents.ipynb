# Getting Started with Python - Overview
Ok, so you are new to python. This notebook isn't meant as a full scale tutorial but as a practical guide on how to begin to use this tool to manipulate and play with data.  Python is a programming language, when measured against the vast array of languages, is fairly easy to learn and powerful in what you can do.  The language is an '*interpreter*' which means it runs right away rather going through the process of preprosessing and compilation. 

On a simple level the **python** language isn't far from your experience with your calculator or in your math class.  Expressions and formulas aren't radically different than what you have seen in you algebra class. There are some syntatic (language specific) characters which the computer needs to undestand what you want it to do but most of these you will learn quickly. There is an old adage in programming: *Computers do exactly what you tell them to do.* Check your answers. If you get an error message, read it carefully.  With time and practice you get better and better at anticipating problems and communicating your intentions. Computers can allow you to process large amounts of data easily.

*There* a number of modules in this folder intended to give you a practical working guide to python.  They are, by no means,  exhaustive but they will get you started. The best way to learn is to practice and use this framework for your classes.
