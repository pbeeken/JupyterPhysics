# Step 4 - Packages (*also known as libraries*)
Python has many powerful built-in features but not everything.  If we included absolutely everything then the language would be come crowded with too many symbols, function names and decorators.  The way around this is to only include those tools you need at any particular time.  These tools are external repositories of code we can 'import' as needed.  One of the tools you have already seen is the  **math** package. 

If we want to take the $sin$ or $cos$ of an angle we need to import the math package and refer to it.  _Never forget that the natural input and output for trig functions is in 'radians' and the default $log$ function is base $e$ (natural log)._ Fortunately there is a function that allows us to convert easily between radians and degrees.  To get package complete list of available mathematical funcions in the **math** package just go to it's [reference site](https://docs.python.org/3/library/math.html).

````
import math

[
 math.sin(math.radians(30.0)),
 math.cos(math.pi/3),
 math.pi, 
 math.log(10.),
 math.log10(10.0)
]
````

## Other Packages
or libraries

While **`math`** is an essential library for basic math calculations using $\sin()$ and $\cos()$ (to name a few) we will be using a library called **`numpy`** which has all the same utility but has some special powers when it comes to handing arrays.  The method and property names are the same but they have the ability to work with arrays of values as well as single (scalar) values.  There is a whole notebook on this library we'll just use this as an example for now.

````
import numpy
# for scalars this gives the same result as the math library
[
 numpy.sin(numpy.radians(30.0)),
 numpy.cos(numpy.pi/3),
 numpy.pi, 
 numpy.log(10.),
 numpy.log10(10.,)
]
````

````
# but it can operate over arrays as well
listOfValues = [numpy.pi / 3, 2 * numpy.pi / 3, 3 * numpy.pi / 3]

# numpy handles this
print(f"{numpy.sin(listOfValues)}")

# math cannot
print(f"{math.sin(listOfValues)}")
````

#### A modelling example using `numpy`


````
import numpy as np  # pretty typical way to import the numpy library and alias it as np for easy reference

# By way of an example lets investigate a small 
# toy car accelerating along a flat surface.

def acceleratingCarData( v0, a ):
  t = np.arange(0.0,10.0,0.5)  # This creates and array of time steps from 0 to just under 10.0s in 0.5s steps
  x = v0*t + (1/2)*a*t**2      # We can calculate the results of a list of numbers without explicit looping
  v = v0 + a*t                 # Next we do the same thing but with velocity
  return np.stack((t,x,v))    # Finally we return the arrays as ordered sets starting with the time, position, and velocity

# assign the data to a single variable that holds all the values
carData = acceleratingCarData( 1.0, -0.1 )

carData.shape  # we have a 3 x 20 array where the columns are the correlated values (t,x,v) 
````

````
carData.transpose()
````

Note a couple of things about the code snippet shown above.  (Noteworthy if you have gone through the python code tutorial)

1. We can create an array of a list of values with a single function call.
0. __numpy__ allows us to do calculations, in parallel, in the same way we can do with variables holding single values.  (Look, ma, no `for` or `while` loops)
0. we can compose arrays (think the columns of a spreadsheet) based on parallel sets of arrays

There is much more we can learn about numpy arrays (a lot more) but for now lets see how this works for plotting.  The following example is intended to show how easy it is to generate a graph from a set of data. 

###<img src="https://www.scipy.org/_static/images/matplotlib_med.png" width="30" /> [matplotlib](https://matplotlib.org/3.1.1/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py) Comprehensive  plotting package

A really, really powerful graphing toolkit that has been ported to many programming languages.  Fortunately we only need a handful of commands to do what we need to do.  Some of these tools are built into pandas (more on this package later).

For this example I will assume you have run the above 'modelling' code snippet under the numpy heading.

````
import matplotlib.pyplot as plt # a standard way to import the basic plotting package and alias it to plt

# let's plot position as a function of time (time is the first row, pos is the second)
plt.scatter(carData[0], carData[1])  # we defined these data sets above.

plt.xlabel("time [s]")  # label our axes with units!
plt.ylabel("pos [m]")
plt.title("Position vs Time")
plt.grid(True)    # lets add a grid so we can evaluate the results.

plt.show()
````

That was easy. The sequence of commands are fairly intuitive as they follow the steps you might if you were doing this by hand.  The advantage to this 'command' line sequence over clicking through menus is twofold:

1. You have a complete record of the steps you took to produce a graph
2. the sequence is easily modified to change and enhance the graphs to better present your data
