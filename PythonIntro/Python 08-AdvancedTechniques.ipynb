## System aliases

You can execute many commands you could use in any terminal environment.  If you've gone through the other modules in this tutorial you have seen this in various guises.  Just put a **Bang!** (yes, a '`!`') everything that follows is passed to the command line.  Most unix commands will work.

```
!ls /bin # list the programs available in this virtual machine
```

```
!ls sample_data/ # list my local directory, compare with what you see when you click on the little folder on the upper left.
```

That `!ls /bin` probably generated a large output. You can select the cell and clear the output by either: 

1. Clicking on the clear output button (x) in the toolbar above the cell (hover over the rectangle); or
2. Right clicking the left gutter of the output area and selecting "Clear output" from the context menu.


Execute any other process using `!` with string interpolation from python variables, and note the result can be assigned to a variable:

```
message = 'Colaboratory is great!'
foo = !echo -e '$message\n$message'
foo
```

## Advanced Printing
You can create all sorts of magical outputs using print.  The most useful is pretty printing to let the system automatically deal with very long output or specially formatted output. The full specification can be found in the [python docs](https://docs.python.org/3/library/string.html#format-specification-mini-language)

```
a = 15
b = 1.2
c = 1.7E-8
d = "string literal"
e = [a,b,c]  # an array

# no qualifiers
print(f"nothing a: {a} b:{b} c:{c} d:{d} e:{e}")
print()

# integer variations in assorted bases and with # output specifier
print(f"various integer formatting a: base10: {a} base16(hexadecimal):{a:#04x} base8(octal):{a:05o} binary:{a:08b}")
print()

# float formats
print(f"various float formats b, c: float b:{b:5.1f} c:{c:5.1f}  exp b:{b:12.2E} c:{c:12.2e}")
print()

# string formats (set which way things get positioned within a declared room)
print(f"various string formats: |{d:20}|{d:>20}|{d:^20}|{d:<20}|")
```
