## Our old pal Newton

Isaac Newton comes along 100 years later and proposes a simple formula for the strength of gravitational pull as a function of distance and the masses of the attracting bodies.

$$F_G = \mathbf{G} \frac{m_1\;m_2}{r^2}$$

The amazing thing about this simple formula is that it made perfect geometrical sense and explained all forms of gravitational attraction from the falling of an apple to the orbit of the moon.
