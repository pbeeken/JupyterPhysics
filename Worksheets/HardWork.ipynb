#Work and Energy

Newton's 3$^{rd}$ law has some remarkable conseqences when it comes to the effect of doing somehing to an object with an imposed force.  
 > *If an object is experiencing a force while moving in a direction that shares some component with that force then the force can be said to 'do work' on the object.*

That was a mouthful.  To unpacking this idea in a more succinct form what we need to tabulate is the direction of motion (displacement) that corresponds to direction the force is acting on an object.  Oh, wait, sounds like a job for ~Super...~ DOT PRODUCT.

$$\mathcal{W}ork = \vec{F} \cdot \vec{x}$$

Almost right.  This, in one simple statement, allows us to calculate the effect of a force's effect on an object being displaced (usually as a result of the force).  What is even more interesting is that the if the force is acting to displace an object and consequently "doing work" on the object. The entity that is providing the force is experiencing exactly the same force in opposition.  The Work that is being done on the object is coming from the entity.  Think of this as an expense that the is being put into the object and being paid by the entity.  Add all the expense up and it stays the same across both objects. Every penny being spent by the entity is taken by the object.

What if the force changes as we travel over the displacement?  Well we can always break the displacement into little tiny pieces and just add up the contributions:

$$\mathcal{W}ork = \int \vec{F} \cdot \mathbb{d}\vec{x}$$

If the force doesn't change over the path then it is constant and the resulting integral is simply what we got above. So this is just a more general way of looking and determining the work done.

Now what do we do with this measurement?  What does 'work' represent?  Lets consider one example: the field free movement of an object.

Let's apply a force $F_A$ to an object.   There are no other forces acting on the object other than the one we supply.  The object will cause the object to accelerate.  After we travel a distance $\ell$ we have done a certain amount of work.

$$\mathcal{W}ork = F_A\;\ell$$

In this simple scenario the force doesn't change over the distance we travelled so we can easily calculate the acceleration.

$$\mathcal{W}ork = m\;a\;\ell$$

After travelling this distance can we get a measure of the change in the speed of the object?  This is a practical outcome of applying an external force. Fortunately we can use one of our old kinematic equations to resolve this: $2\;a\;\ell = v_f^2-v_i^2$ we result in the following:

$$\mathcal{W}ork = \frac{1}{2}m\;\left(v_f^2-v_i^2\right) = \frac{1}{2}m\;v_f^2 - \frac{1}{2}m\;v_i^2$$

We can define this combined value, $\color{red}{\frac{1}{2}m\;v^2}$ as the **Kinetic Energy**.

Would this still be true if the force varied over the distance?  Lets try this out:
$$
\begin{align}
\mathcal{W}ork &= \int_0^\ell F(x)  \mathbb{d}x \\
               &=  m \int_0^\ell a(x)  \mathbb{d}x \\
               &=  m \int_0^\ell \left(\frac{\mathbb{d}}{\mathbb{dt}} v\right) \mathbb{d}x 
\end{align}
$$
We can, as we did with differential equations, play the derivative is a fraction card and move the differentials so we are operating on the same on the same elements.  Notice that as we do we reset where the positional dependant velocity is declared.
$$
\begin{align}
\mathcal{W}ork &= m \int_0^\ell \mathbb{d}v \frac{\mathbb{d}x}{\mathbb{dt}}\\
\mathcal{W}ork &= m \int_0^\ell \frac{\mathbb{d}x}{\mathbb{dt}} \mathbb{d}v \\
\mathcal{W}ork &= m \int_0^\ell v(x) \mathbb{d}v \\
\mathcal{W}ork &= m \frac{v(x)^2}{2}\Bigg\rvert^\ell_0 \\
\mathcal{W}ork &= \frac{1}{2}m\;v(\ell)^2 - \frac{1}{2}m\;v(0)^2 \\
\end{align}
$$

The significance of this is hard to understate. The work done on a system can be manifested as a simple differnce between the before and after conditions of the system. The details of the force do not matter.

### A pebble for your thoughts
A $2 kg$ pebble (pebble? this is a rock! or maybe it's Pebbles the Flinstone's child? I digress.) is moved along a horizontal surface with a force that varies as the pebble's position as outlined by the following graph:

```
#@title Plot of force acting on "pebble"
def forceA(x):
  def f1(x):
    return 1/2 * x

  def f2(x):
    return 1
  
  def f3(x):
    return 1 - 9/10 * x

  def f4(x):
    return -4/5
  
  def f5(x):
    return -4/5 + 4/5 * x

  func = 5 * np.piecewise(x,
                          [x<=2.0         , (x>2.0) & (x<=4) , (x>4.0) & (x<=6.0), (x>6.0) & (x<=7.0), x>7.0            ],
                          [lambda x: f1(x), lambda x: f2(x-2), lambda x: f3(x-4) , lambda x: f4(x-6) , lambda x: f5(x-7)])
  return func


x = np.arange(0.0, 8.0, 0.1)   # 0.0 and 8.0 m
# begin the plot process...
plt.style.use('classic')

# plot the points connecting with a smooth line  
# c is the color, you can make multiple plots
# linewidth is the thickness
# alpha is the transparency (1:opaque, 0:invisible)
plt.plot(x, forceA(x), c="b", label="force", linewidth=3, alpha=0.5)

# label the axes and graph
plt.xlabel("position [m]")
plt.ylabel("F [N]")
plt.title("Force on a Pebble")

# put up a legend
plt.legend(loc="upper left")  # has to appear after we plot.
plt.grid(True)
plt.ylim((-5,6))
# show it.
plt.show()
```
