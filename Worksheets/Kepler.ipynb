## Kepler's Laws


In Aristotle's time the heavens were perfect. Stars and planet were perfect spheres created by the divine and except for the 'wanderers' also known as œÄŒªŒ±ŒΩŒÆœÑŒ∑œÇ or planets the heavens remained immutable, never changing.

Their orbits were perfect circles... which is almost true if you don't look carefully, which Brahe and Kepler did and things weren't quite so perfect.  Add to this was the discovery of that pesky itallian Gallileo who trained an optical telescope at the moon and noticed it was anything but a perfect sphere. The moon had mountains and valleys and holes. It was well on its way to loosing the Aristotle Housekeeping Seal of Approval. I'll summarize the key ideas of Kepler's discoveries:

 1. <font color="cornflower"> Planets move in ellipses (a circle is just a special case of ellipse where the short axis (semi-minor axis) equals the long axis (semi-major axis) with the sun at the center (Newton would later change this to the largest attracting body)
 </font>

 2. <font color="cornflower"> The timing of the orbital motion is such that as the planet turns in its orbit the planets sweep out triangle like shapes such that the area is the same for similar time intervals. In other words planets would speed up and slow down as they travelled around the central attactor (the sun, for example) that for any fixed interval of time (a day, a week, even an hour) the area created using an imaginary line from the planet to the sun at the beginning and ending point would be the same.
</font>

 2. <font color="cornflower">The overall orbital period squared of any orbit scales as the cube of the semi-major axis.
</font>

These basic ideas were revolutionary in the early 17th century.  Newton simply reasoning that the rules governing the falling of an apple must also explain the orbit of the moon came up with a model (and a mathematical tool to prove) that justified these three laws.

```
#@title Kepler's 3rd Law Demonstration
plt.xkcd()

# This seems a little weird but is necessary because numpy doesn't handle units perfectly.

# This first line guarantees that all the axis information is 
# measured in AU before we just store it in an array
radii = [(planetData[p]['semimajoraxis'].to(SI.astronomical_unit).magnitude) for p in planetData]
# This simply puts the units back in after we recreate the numpy array.
radii = np.array(radii)*SI.astronomical_unit # numpy arrays discard the units so we re-establish it

# Same trick as above, we first make sure all the periods are measured in days...
period = [(planetData[p]['orbitperiod'].to(SI.day).magnitude) for p in planetData]
# ... then re-establish the units after we store them in the array.
period = np.array(period)*SI.day # numpy arrays discard the units so we re-establish it

plt.scatter(period**2, radii**3)

plt.title("Kepler's 3rd Law")
plt.ylabel("$SemiMajor Axis^3 [Au^3]$")
plt.xlabel("$Orbital Period^2 [yr^2]$")

# the way to determine a proportional relationship coefficient (we aren't looking for a linear relationship)
GmsOverFourPiSq = np.average(radii**3 / period**2) # get a slope the intercept has to be 0.  The slope should be G*Ms/4ùùÖ¬≤

plt.plot(period**2, GmsOverFourPiSq * period**2, c='c', alpha=0.4)
# resetablish units

Gms = 4 * np.pi**2 * GmsOverFourPiSq #* (SI.astronomical_unit**3) / (SI.day**2) # restore the units
Gexp = Gms.to_base_units() / starData['Sun']['mass']

plt.xlim(0.0)
plt.ylim(0.0)

# Because of the units we will get a warning message (the plotting routines create
# their own arrays internally.) This is OK because everything is scaled properly.
plt.show()

# Finally some comparions to our solar system vs lab measurements...
print(f"Newton's Universal Grav Constant: {Gc.to_base_units():10.5g}")
print(f"Univ Grav Constant ala Kepler: {Gexp:10.5g}")

print(f"{100*(abs(Gc - Gexp)/Gc).magnitude:4.3}% deviation")
print("I guess our solar system follows Newton's and Kepler's Laws!")
```
