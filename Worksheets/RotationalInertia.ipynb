# Rotational Inertia


<img alt="rolling racers" src="https://upload.wikimedia.org/wikipedia/commons/2/2e/Rolling_Racers_-_Moment_of_inertia.gif" align="right" width="300px"/>Objects that rotate have inertia that affect how fast the angular speed of a spinning object can change.  This is exactly analogous to the inertia we associated with speeding or slowing an object in a linear motion.  The difference is in how we caclulate it.

Inertia for linear motion only considers the total mass. For rotational inertia we need to know something about the distribution of mass about the axis of spin.

## Single Mass

A single mass $m$ spinning around an axis at a distance $r$ has a moment of inertia that equates to $\mathbb{I} = m\;r^2$. But what if we have a bunch of masses $m_i$ all spinning about the same axis but at different radii $r_i$?  As it is with regular mass in linear inertia calculations we should be able to add the contributions up.

$$\mathbb{I} = \sum_{i=1}^N m_i\;r_i^2\tag{1.3}$$

If we are dealing with continuous distributions of mass we would express this as tiny little bits of mass $\mathbb{d}m$ as an integral
$$\mathbb{I}=\int_0^R r^2 dm\tag{1.4}$$

These calculations also tells us how to build moments of inertia for solid materials as well.  It is useful to go through one of these derivations to understand the steps but not dwell too long on it as we will look up the result for most simple geometries.

### Derivation of the moment of Inertial for a disk

<img alt="spinning disk" src="https://upload.wikimedia.org/wikipedia/commons/f/fc/Spinning-disk.svg" align="right" width="150px"/>

We want to calculate the moment of inertia $\mathbb{I}$ of a solid disk of radius $R$ and total mass $M$ $^\dagger$.  We start with the summation $1.3$ above but what we want to do is to sum up tiny little mass bits $\mathbb{d}m$ as we count them up going from the center out to the edge. This is tricky because as we move outward we have more mass so we need to calculate the mass as function of postion as we move outward.  We need a relationship between mass and radius so we can complete the integral in $1.4$. We need a forumla so we can substitute for the differential $\mathbb{d}m$.

We need something that stays constant as we do this: **We presume we have uniform density** $\rho = \frac{M}{volume}$ where volume can for a cylinder can be used: $volume = \pi R^2 h$. With this we can work out the relationship to mass and radius as (chain rule, effectively)
$$
\begin{align}
m_{ass}(r) &= \rho &\pi r^2 h \\
    m(r)    &= \frac{M}{\pi R^2 h} &\pi r^2 h \\
    m    &= \frac{M}{R^2} &r^2\\
\mathbb{d}m &=2\frac{M}{R^2} & r\;dr  \tag{1.5} 

\end{align}
$$
So the integral $1.4$ becomes (after substitution)
$$\require{cancel}
\begin{align}
\mathbb{I}&=&\int_0^R r^2 dm \\
\mathbb{I}&=2\frac{M}{R^2} &\int_0^R r^2\;r\;dr\\
\mathbb{I}&=\cancel{\color{darkred}2}\frac{M}{R^2} &\frac{r^4}{\cancel{\color{darkred}4}2}|_0^R\\
\mathbb{I}&=\frac{1}{2}\frac{M}{\cancel{\color{darkred}{R^2}}} &R^{\cancel{\color{darkred}4}2}\\
\mathbb{I}&=\frac{1}{2}M\;R^2&\tag{1.6}
\end{align}
$$

Notice how the thickness of the cylinder along the axis plays no role in the calculation but does show us that if we strap two side by side along the same spin axis we need only add the rotational inertia of the first cylinder, $\mathbb{I_1}$, to the rotational inertia of the second, $\mathbb{I_2}$

### Derivation of the moment of Inertial for a rod spinning about the middle

Here we will do the same thing but with a 'thin' rod of length $L$ and mass $M$ spinning about its center.

<img alt="spinning rod" src="https://upload.wikimedia.org/wikipedia/commons/thumb/b/b4/Moment_of_inertia_rod_center.svg/170px-Moment_of_inertia_rod_center.svg.png" align="right" width="150px"/>

Again we will resort to the integral form of the moment of inertia $1.4$ but our derivation of the mass as a function of distance away from the center is much easier.  The density of the thin rod is simply $\rho = \frac{M}{A\;L}$ where the cross sectional area of the rod is considered inconsequential compared to the length so we don't care if it circular, square, or any other shape. We only presume the cross section to be uniform along the length.

$$
\require{cancel}
\begin{align}
m_{ass}(r) &= \rho\; x\; A \\
        &= \frac{M}{\cancel{\color{darkred}A}\;L} \; x\; \cancel{\color{darkred}A}\\
\mathbb{d}m &=\frac{M}{L}\;dx \tag{1.7}
\end{align}
$$
In addition to the substituttion we also change the limits of the integral from $0$ to $R$ to one end to another or from $-\frac{L}{2}$ to $\frac{L}{2}$. Now the integral $1.4$ becomes:
$$\require{cancel}
\begin{align}
\mathbb{I}&=&\int_{-L/2}^{L/2} x^2 dm \\
\mathbb{I}&=\frac{M}{L}&\int_{-L/2}^{L/2} x^2\;dx\\
\mathbb{I}&=\frac{M}{L} &\frac{x^3}{3}|_{-L/2}^{L/2}\\
\mathbb{I}&=\frac{1}{3}\frac{M}{L} &\left(L/2\right)^3 - \left(-L/2\right)^3\\
\mathbb{I}&=\frac{1}{3}\frac{M}{\cancel{\color{darkred}L}} &\left(L^{\cancel{\color{darkred}3}2}/4\right)\\
\mathbb{I}&=\frac{1}{12} M\;L^2 \tag{1.8}\\
\end{align}
$$
Hey the dimenstions/units worked out! But what if it is spinning about the end?  We don't need do anything except change the limits of the integral above from 0 to L.  Is the formula the same? Does it make $\mathbb{I}$ larger or smaller to move the axis to the end?

#### Parallel Axis Theorem

Now we want to consider what happens if we want to shift the axis on one of the objects.  Above we have an integral calculation of the moment of inertia of a rod about its center and about one end. Each derived independantly.  Is there a way we could have done this without doing an integral once we had one moment of inertia spinning about its center of mass?

**Take an object of mass $M$ with a moment of inertia about its center of mass, $\mathbb{I}_{cm}$ and displace the spin axis parallel to the first by a distance $d$, what is the resulting moment of inertia, $\mathbb{I}_{d}$?**

$$\mathbb{I}_{d} = \mathbb{I}_{cm} + M\;d^2\tag{1.10}$$

That seems simple. Notice some important features of this expression. The first, most important one, is that we don't need to know anything about the details of the shape of the object or how we got the original moment of inertia.  The second facet is that when d gets really large (larger than the size of the object) the center of mass moment of inertia, $\mathbb{I}_{cm}$, becomes negligible w.r.t. $M\;d^2$ so the formula looks more like a point mass being swung around an axis of radius r (*take a look at the first expression we had for a moment of inertia of a lone mass*.)

## Combining complicated things

#### Thick Ring

The moment of inertia of a hoop is relaively simple to work out, all the mass $M$ is concentrated in a thin ring at some radius $R$
  - Thin ring about center: $\mathbb{I}_{cm} = M\;R^2$
  - Disk about center: $\mathbb{I}_{cm} = \frac{1}{2}M\;R^2$ 
  - Rod about middle: $\mathbb{I}_{cm} = \frac{1}{12}M\;L^2$
  - Solid Sphere about middle (may need it later): $\mathbb{I}_{cm} = \frac{2}{5}M\;R^2$
